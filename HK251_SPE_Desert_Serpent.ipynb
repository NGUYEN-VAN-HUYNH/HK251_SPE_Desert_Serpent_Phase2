# 2.Thiết kế hệ thống

### 1. Define Goal and System


#### __Mục tiêu hệ thống__

Hệ thống cần đánh giá là mô hình hàng đợi của một khu vui chơi giải trí, nơi khách hàng được cung
cấp các dịch vụ giải trí theo nhiều hình thức xếp hàng đa dạng. Những mục tiêu cơ bản của dự án
đánh giá hiệu năng như sau:

- Mục tiêu chính là cần đánh giá hiệu năng hoạt động của các loại hàng đợi trong hệ thống có
ảnh hưởng như thế nào đến thời gian đáp ứng (response time) trung bình của toàn khu vui chơi.
Sau đó, tìm ra điểm tắt nghẽn để đưa ra cải tiến đáp ứng được nhu cầu của khách hàng.
- Hệ thống cũng cần đảm bảo phải tận dụng tốt nhất các tài nguyên sẵn nhằm để tối đa hóa lợi
nhuận, đồng thời giúp giảm thiểu chi phí đầu tư phát triển.
- Ngoài ra, hệ thống cũng cung cấp một số dịch vụ ưu tiên cho khách hàng VIP nên cần đánh giá
xem cơ chế ưu tiên có hoạt động đúng với mong muốn hay không. Cụ thể là so sánh thời gian
đáp ứng giữa khách thường và khách VIP.

#### __Định nghĩa hệ thống__
Khu vui chơi giải trí được thiết kế bao gồm 6 khu vực với chức năng và cách thức phục vụ khách hàng
khác nhau. Ở mỗi khu vực có một hoặc nhiều mô hình hàng đợi đa dạng, được thiết kế dựa vào đặc
tính của từng khu vực.
1. Khu vực mua vé: Nơi này gồm hai thành phần chính là quầy mua vé thường và quầy mua vé
VIP. Mỗi quầy vé có dạng hàng đợi M/M/1.
2. Khu vực check in: Có một cổng (server) để check in, trước cổng có 2 loại hàng đợi dành cho vé
thường hoặc VIP. Tất cả khách hàng VIP sẽ được ưu tiên vào cổng trước ở hàng đợi ưu tiên. Sau
đó, khách hàng ở hàng đợi thường sẽ được check in. Mô hình hàng đợi ở đây tuân theo M/G/1
3. Khu vực trò chơi: Khu vực này cũng sẽ ưu tiên phục vụ cho khách VIP nhanh hơn. Gồm hai
loại hàng đợi, M/M/c dành cho vé thường và M/M/1 cho vé VIP.
4. Khu vực ăn uống: Không phân biệt loại vé, mọi người đều xếp hàng chung trong 1 hàng đợi
M/M/c để chờ đến lượt được phục vụ thức ăn.
5. Khu vực triễn lãm: Đây là khu vực duy nhất chỉ dành riêng cho khách có vé VIP. Nơi xét vé có
mô hình hàng đợi dạng M/M/1.
6. Khu vực check out: Có 1 hàng đợi M/M/c để phục vụ khách hàng check out rời khỏi hệ thống
khu vui chơi giải trí.

#### __Giả định của hệ thống__
1. Khách hàng mới đến nếu đã có vé (online) VIP hoặc thường sẽ đi thẳng đến khu vực check in mà không cần qua Khu vực mua vé.
2. Khách hàng sau khi check in phải qua ít nhất một khu vực để được đến khu vực check out.
Trình tự khách hàng có thể đến và rời khỏi từng khu vực là: mua vé → check in → tham gia
các hoạt động trò chơi, ăn uống, triểm lãm không cần theo thứ tự và không giới hạn số lần → check out.
3. Nếu tại khu vực có phục vụ hàng đợi VIP, khách hàng thường sẽ không được dùng hàng đợi VIP, và để đơn giản thì khách hàng VIP cũng không vào hàng đợi thường, Tại các khu vực còn lại thì cả khách hàng VIP và khách hàng thường đợi tại một hàng đợi dùng chung.

### 4. List System Parameters



Dựa trên các định nghĩa và giả định ban đầu của hệ thống, các tham số ảnh hưởng đến hệ thống như sau:
* **$\lambda $ :** Tốc độ đến tổng thể Đây là tốc độ mà khách hàng mới đến toàn bộ hệ thống.
* **$C_i$:** Tương ứng với **số cổng** hoặc **máy chủ** ($c$) tại từng khu vực $i$. Ví dụ: $C_{CheckIn}$ là số cổng Check In.
* **$\mu_i$ :** Tốc độ phục vụ .Biểu thị số khách hàng **tối đa** mà một cổng/máy chủ $i$ có thể phục vụ trong một đơn vị thời gian.
* **$p_{ij}$:** Xác suất một khách hàng **chuyển từ khu vực $i$ đến khu vực $j$** sau khi được phục vụ tại $i$.

### 6. Evaluation Technique

Do Hệ thống Khu Vui Chơi Giải Trí đang ở giai đoạn **thiết kế mô hình**, nhóm quyết định kết hợp hai kỹ thuật mô hình hóa:

* **Kỹ thuật chính:** Mô phỏng rời rạc theo sự kiện (**Discrete Event Simulation – DES**).
* **Kỹ thuật hỗ trợ:** Mô hình phân tích dựa trên **Lý thuyết Hàng đợi (Queueing Theory)**.

Nhóm áp dụng chiến lược sử dụng kỹ thuật tuần tự để tối ưu hóa thời gian và nguồn lực:

##### A. Giai đoạn 1: Mô hình Phân tích (Analytical Modeling)

* **Mục đích:** Thu hẹp phạm vi tham số đầu vào.
* **Ứng dụng:**
Sử dụng công thức Lý thuyết Hàng đợi (M/M/1, M/M/c) để tính toán các giới hạn trên của **độ tận dụng tài nguyên ($\rho_i$)** tại các khu vực quan trọng (ví dụ: Check Out, Trò chơi).

#### B. Giai đoạn 2: Mô hình Mô phỏng (Simulation)

* **Mục đích:** Nghiên cứu **hiệu ứng tương tác (interactions)** giữa các nhân tố và đánh giá hiệu năng trong phạm vi đã được thu hẹp.
* **Lý do cần mô phỏng:** Mô hình hàng đợi toàn bộ hệ thống là một Mạng Hàng đợi phức tạp có chứa cơ chế ưu tiên, khó có thể giải quyết chính xác bằng công thức phân tích.


### 7. Select Workload




#### Định nghĩa và Mục tiêu của Workload

**Khối lượng Công việc (Workload)** là tập hợp các yêu cầu dịch vụ do khách hàng đưa ra cho Hệ thống đang được thử nghiệm (SUT).

* **SUT (System Under Test):** Mô hình hàng đợi của Khu Vui Chơi Giải Trí.
* **Workload:** Các yêu cầu của khách hàng để sử dụng dịch vụ trong mô hình.

##### A. Thành phần Workload

Dựa trên mục tiêu đánh giá cơ chế ưu tiên, khách hàng được phân loại thành:

* **Khách Hàng Thường (Regular Customer - Reg)**
* **Khách Hàng VIP (VIP Customer - VIP)**

##### B. Hình thức Workload được chọn

Vì kỹ thuật đánh giá chính được chọn là **Mô phỏng (Simulation)**, nhóm sẽ sử dụng **Workload Tổng hợp (Synthetic Workload)**. Workload này được mô tả bằng phân bố xác suất hoàn chỉnh của nhu cầu tài nguyên và xác suất chuyển đổi. Lý do chọn loại Workload này:
* **Lặp lại (Repeatability):** Đảm bảo các thí nghiệm có thể được lặp lại với các tải tương đương trong điều kiện được kiểm soát.
* **Kiểm soát các Nhân tố (Factors):** Cho phép thay đổi có hệ thống các tham số Workload.


#### Các Nhân tố Workload được Chọn để Nghiên cứu

Các tham số có tác động lớn được chọn làm **nhân tố (factors)** để thay đổi trong quá trình mô phỏng.

| STT | Nhân tố Workload | Ký hiệu | Phạm vi Thử nghiệm (Levels) | Mục tiêu Nghiên cứu |
| :---: | :--- | :---: | :--- | :--- |
| **1** | **Tốc độ đến Khách Thường** | $\lambda_{\text{reg}}$ | **[60, 100, 150] khách/giờ** | Nghiên cứu hiệu năng (đặc biệt là $E[R]$) dưới các mức tải khác nhau (ngày thường, cuối tuần, lễ hội). |
| **2** | **Tỷ lệ Khách VIP/Thường** | $\lambda_{\text{VIP}}/\lambda_{\text{reg}}$ | **[1 : 10, 1 : 6, 1 : 4]** | Đánh giá khi nào cơ chế ưu tiên VIP trở nên đắt đỏ (cản trở) đối với Khách Thường. |


#### Kết luận về Workload

Workload được phát biểu dưới dạng **Mô hình Thống kê (Statistical Workload Model)**, bao gồm:

* **Mô hình đến Poisson** với $\lambda_{\text{reg}}, \lambda_{\text{VIP}}$ được kiểm soát.
* **Mô hình chuyển đổi Markov** ($p_{ij}$) dựa trên FSM của khách hàng.
* **Phân bố thời gian phục vụ** tại mỗi khu vực (giả định M/M/c hoặc M/M/1).

Workload này sẽ được đưa vào mô hình mô phỏng để thu thập các chỉ số hiệu năng như **Thời gian đáp ứng trung bình $E[R]_{\text{VIP}}$ và $E[R]_{\text{Reg}}$**.

Sơ đồ lớp (Class Diagram) trên  mô tả kiến trúc phần mềm của hệ thống mô phỏng sự kiện rời rạc (Discrete Event Simulation), được thiết kế theo mô hình tập trung vào bộ điều khiển (SimulationModule). Hệ thống bao gồm các thành phần chính sau:

1. Khối Điều khiển Trung tâm (Controller)
SimulationModule: Đóng vai trò là "bộ não" điều phối toàn bộ quá trình mô phỏng. Lớp này chịu trách nhiệm:

* Khởi tạo môi trường mô phỏng (run_simulation).


*   Điều phối luồng khách hàng (customer_generator sinh ra khách và customer_process xử lý hành vi khách).
*  Tính toán lộ trình di chuyển động (generate_dynamic_path) dựa trên xác suất chuyển trạng thái.
*  Thực hiện phân tích dữ liệu hàng đợi sau khi mô phỏng kết thúc (empirical_queue_analysis).
2. Khối Dữ liệu & Thực thể (Entities)
* CustomerData: Đại diện cho thực thể khách hàng trong hệ thống. Lớp này lưu trữ các thuộc tính quan trọng như mã định danh (id), loại khách (type), thời gian đến, và lộ trình di chuyển.

* CustomerType: Một Enumeration định nghĩa hai loại khách hàng chính trong hệ thống là REG (Thường) và VIP, giúp phân loại quyền ưu tiên xử lý.

3. Khối Tài nguyên (Resources)
* ResourceWrapper: Là lớp bao đóng (wrapper) cho thư viện simpy.PriorityResource. Nó không chỉ quản lý hàng đợi (queue) mà còn mở rộng thêm các thuộc tính nghiệp vụ cụ thể như: thời gian phục vụ trung bình (service_time_mean), cấu hình ưu tiên cho khách VIP (vip_priority), và sức chứa (capacity). Các đối tượng này được khởi tạo dựa trên thông số từ Config.

4. Khối Cấu hình & Tiện ích (Configuration & Utilities)
* Config: Chứa toàn bộ các tham số tĩnh của hệ thống như cấu hình khu vực (AREAS_CONFIG), sức chứa công viên, và ma trận xác suất chuyển tiếp. Việc tách biệt này giúp dễ dàng thay đổi tham số mô phỏng mà không cần sửa đổi logic code.

* Logger: Chịu trách nhiệm ghi lại toàn bộ sự kiện diễn ra trong quá trình mô phỏng (khách đến, xếp hàng, phục vụ xong) ra file CSV và màn hình console. Lớp này giúp tách biệt logic ghi log khỏi logic xử lý nghiệp vụ.

5. Tương tác với Thư viện Ngoài (External Libraries)
* Hệ thống tương tác chặt chẽ với các thư viện bên thứ ba: SimPy (cung cấp lõi mô phỏng sự kiện), Pandas/NumPy (xử lý tính toán thống kê), và CSV (xuất dữ liệu).

**=>** Kiến trúc này đảm bảo tính tách biệt giữa dữ liệu (Customer, Config), logic xử lý (SimulationModule), và tài nguyên hệ thống (ResourceWrapper), giúp hệ thống dễ dàng mở rộng và bảo trì khi cần thay đổi quy trình phục vụ hoặc thêm loại khách hàng mới.

* Chủ khu vui chơi nhận thấy hiện hệ thống chỉ đáp ứng được lượng khách đến những ngày trong tuần. Với những ngày cuối tuần hay ngày lễ hội thì hệ thống sẽ bị quá tải. Để giải quyết vấn đề đó, chủ khu vui chơi đã cải tiến hệ thống bằng cách cải tiến tốc độ phục vụ trong trường hợp cuối tuần (100 khách thường/giờ và 10 khách vip/giờ) và giữ nguyên số máy chủ hiện tại ở các khu vực (giả sử hiện tại mỗi khu vực đều có 1 máy chủ). Lúc này nhóm em đóng vai trò như chủ khu vui chơi.

* Để làm được điều này, sử dụng phương pháp xác định tham số mô phỏng (Biện luận thiết kế) xác định bộ tham số cấu hình $(c, \mu)$ tối ưu cho hệ thống nhằm giải quyết bài toán tắc nghẽn, áp dụng phương pháp "Phân tích dòng chảy tới hạn" (Critical Flow Analysis).Quy trình tính toán được thực hiện dựa trên giả định hệ thống hoạt động tại biên của sự ổn định (Boundary of Stability). Cụ thể, ta thiết lập hệ phương trình cân bằng dòng dựa trên nguyên tắc: Tại trạng thái bão hòa, năng lực phục vụ của trạm phải đáp ứng chính xác lượng khách trung bình đến trạm đó.Phương trình thiết kế được xác định như sau:

  $$c_i \cdot \mu_i = \lambda_i$$

  Trong đó:

    * $\lambda_i$: Tốc độ đến tổng hợp tại trạm $i$ (bao gồm cả khách mới và khách quay lại từ các trạm khác), được tính toán thông qua hệ phương trình cân bằng dòng Jackson Network.

    * $c_i \cdot \mu_i$: Tổng năng lực phục vụ (Total Service Capacity) của trạm $i$.

  Lý do lựa chọn phương pháp này:

    * Xác định Cận dưới (Lower Bound Determination): Việc giải phương trình với dấu bằng ($\rho = 1$) giúp tìm ra giá trị $\mu_{min}$ (tốc độ phục vụ tối thiểu). Đây là ngưỡng sàn kỹ thuật; bất kỳ giá trị $\mu$ nào nhỏ hơn giá trị này đều chắc chắn gây ra sự sụp đổ hệ thống (Queue Blow-up).

    * Tối ưu hóa tài nguyên (Resource Optimization): Trong bài toán kinh tế, việc dư thừa năng lực phục vụ ($\mu \gg \lambda$) gây lãng phí tài nguyên. Việc tìm ra điểm cân bằng $\mu \approx \lambda$ giúp xác định mức đầu tư tài nguyên sát với nhu cầu thực tế nhất.

    * Cơ sở cho mô phỏng thực tế: Giá trị $\mu$ tìm được từ hệ phương trình này sẽ đóng vai trò là Baseline (Mức cơ sở). Trong quá trình thực thi mô phỏng (Phase 2), ta sẽ thiết lập tốc độ phục vụ thực tế $\mu_{real} \ge \mu_{calc}$ (hoặc $c$ tăng lên) để tạo ra một biên an toàn (Safety Margin), đảm bảo $\rho < 1$ để hệ thống đạt trạng thái ổn định (Steady State).

* Bước 1: Giải quyết tắc nghẽn tại cổng vào (Check-in Queue), khu vực Check-in là điểm nhập duy nhất của hệ thống, chịu tải trực tiếp từ lưu lượng khách đến bên ngoài ($\gamma$).Trong kịch bản tải cao (cuối tuần), lưu lượng đến dự kiến là:$$\lambda_{CheckIn} = \lambda_{Arrival} = \lambda_{Reg} + \lambda_{VIP} = 100 + 10 = 110 \text{ (khách/giờ)}$$Để hàng đợi không bị bùng nổ ngay tại cổng vào, năng lực phục vụ tổng thể của Check-in phải được thiết lập $\mu_{CheckIn} = 110$ khách/giờ

* Bước 2: Giải quyết tắc nghẽn tại các khu vực nội bộ (Service Queues). Sau khi qua Check-in, khách hàng phân tán vào các khu vực dịch vụ: Trò chơi (Entertainment), Ăn uống (F&B), và Triển lãm (Exhibition). Tại đây, tốc độ đến ($\lambda_j$) không chỉ phụ thuộc vào khách từ Check-in mà còn bao gồm lượng khách luân chuyển giữa các khu vực (feedback flow).Áp dụng phương trình cân bằng dòng (Flow Balance Equations) cho trạng thái ổn định:$$\lambda_j = \lambda_{CheckIn} \cdot P_{CheckIn \to j} + \sum_{k \in Areas} \lambda_k \cdot P_{k \to j}$$Trong đó $P_{i \to j}$ là xác suất chuyển từ khu vực $i$ sang khu vực $j$ (được xác định từ bảng xác suất di chuyển). Ta sẽ thiết lập và giải hệ phương trình tuyến tính này để tìm ra $\lambda$ chính xác tại từng nút (Trò chơi, Ăn uống, Triển lãm). Từ đó, xác định cặp tham số $(c, \mu)$ cho từng khu vực sao cho năng lực phục vụ tối thiểu bằng $\lambda$ tính được.

  Với giả định số lượng máy chủ tại mỗi nút $c=1$ để tính toán năng lực phục vụ tối thiểu, ta có phương trình cân bằng: $\lambda_{i} = \mu_{i}$. Dưới đây là hệ phương trình chi tiết cho từng khu vực:

    1. Tại khu vực Ăn uống (Food & Beverage)Lưu lượng đến khu vực ăn uống là tổng hợp từ tất cả các nguồn: khách mới từ Check-in (cả Thường và VIP) và khách luân chuyển từ các khu vui chơi khác.$$\lambda_{AnUong} = \lambda_{CheckIn}^{Reg} \cdot P_{CI \to AU}^{Reg} + \lambda_{CheckIn}^{VIP} \cdot P_{CI \to AU}^{VIP} + \lambda_{TrienLam} \cdot P_{TL \to AU} + \lambda_{TroChoi}^{Reg} \cdot P_{TC \to AU}^{Reg} + \lambda_{TroChoi}^{VIP} \cdot P_{TC \to AU}^{VIP}$$Áp dụng điều kiện $\lambda = \mu$ (với $c=1$) và thay số:$$\mu_{AnUong} = 100 \cdot 0.3 + 10 \cdot 0.2 + \mu_{TrienLam} \cdot 0.25 + \mu_{TroChoi}^{Reg} \cdot \frac{1}{3} + \mu_{TroChoi}^{VIP} \cdot 0.25 \quad (1)$$

    2. Tại khu vực Triển lãm (Exhibition)Lưu lượng đến khu triển lãm bao gồm khách VIP từ Check-in và khách luân chuyển từ khu Ăn uống hoặc Trò chơi VIP.$$\lambda_{TrienLam} = \lambda_{CheckIn}^{VIP} \cdot P_{CI \to TL} + \lambda_{AnUong} \cdot P_{AU \to TL} + \lambda_{TroChoi}^{VIP} \cdot P_{TC \to TL}$$Áp dụng điều kiện $\lambda = \mu$ (với $c=1$) và thay số:$$\mu_{TrienLam} = 10 \cdot 0.3 + \mu_{AnUong} \cdot 0.25 + \mu_{TroChoi}^{VIP} \cdot 0.25 \quad (2)$$

    3. Tại khu vực Trò chơi Thường (Regular Games)Lưu lượng đến khu trò chơi thường bao gồm khách Thường từ Check-in và khách quay lại từ khu Ăn uống.$$\lambda_{TroChoi}^{Reg} = \lambda_{CheckIn}^{Reg} \cdot P_{CI \to TC}^{Reg} + \lambda_{AnUong} \cdot P_{AU \to TC}^{Reg}$$Áp dụng điều kiện $\lambda = \mu$ (với $c=1$) và thay số:$$\mu_{TroChoi}^{Reg} = 100 \cdot 0.7 + \mu_{AnUong} \cdot \frac{1}{3} \quad (3)$$

    4. Tại khu vực Trò chơi VIP (VIP Games)Lưu lượng đến khu trò chơi VIP bao gồm khách VIP từ Check-in và khách luân chuyển nội bộ giữa các khu vực VIP.$$\lambda_{TroChoi}^{VIP} = \lambda_{CheckIn}^{VIP} \cdot P_{CI \to TC}^{VIP} + \lambda_{AnUong} \cdot P_{AU \to TC}^{VIP} + \lambda_{TrienLam} \cdot P_{TL \to TC}$$Áp dụng điều kiện $\lambda = \mu$ (với $c=1$) và thay số:$$\mu_{TroChoi}^{VIP} = 10 \cdot 0.5 + \mu_{AnUong} \cdot 0.25 + \mu_{TrienLam} \cdot 0.25 \quad (4)$$

  Giải hệ phương trình tuyến tính gồm 4 phương trình (1), (2), (3) và (4) ở trên, ta thu được giá trị năng lực phục vụ tối thiểu ($\mu_{min}$) tại các nút mạng hàng đợi để đảm bảo điều kiện ổn định ($\rho \le 1$) như sau:
    * Khu vực Triển lãm: $\mu_{TrienLam} \approx 32$ (khách/giờ)
    * Khu vực Trò chơi Thường: $\mu_{TroChoi}^{Reg} \approx 97$ (khách/giờ)
    * Khu vực Trò chơi VIP: $\mu_{TroChoi}^{VIP} \approx 33$ (khách/giờ)
    * Khu vực Ăn uống: $\mu_{AnUong} \approx 81$ (khách/giờ)


* Bước 3: Xác định năng lực phục vụ tại cổng ra (Khu vực Check-out)
  * Khác với các khu vực dịch vụ nội bộ, khu vực Check-out đóng vai trò là "van xả" cuối cùng của hệ thống. Năng lực phục vụ tại đây không chỉ phụ thuộc vào lưu lượng dòng khách hiện tại mà còn bị ràng buộc bởi Sức chứa thiết kế tối đa (Maximum Design Capacity) của công viên.
  * Theo giả định ban đầu, cơ sở hạ tầng vật lý của khu vui chơi được xây dựng để đáp ứng tối đa 200 khách hoạt động đồng thời. Để đảm bảo an toàn hệ thống và tránh hiện tượng tràn khách (overflow) khi đạt tải đỉnh, tốc độ giải phóng khách tại cổng ra phải tương đương với sức chứa tối đa này.
  * Quyết định thiết kế: Ta thiết lập tốc độ phục vụ tại Check-out là:$$\mu_{CheckOut} = 200 \text{ (khách/giờ)}$$
  * Giá trị này lớn hơn đáng kể so với lưu lượng ra thực tế tại trạng thái ổn định ($\lambda_{out} \approx 110$). Việc thiết lập $\mu_{CheckOut} \gg \lambda_{out}$ tạo ra một biên an toàn lớn (Safety Margin), đảm bảo rằng ngay cả khi có sự gia tăng đột biến lượng khách muốn ra về cùng lúc (bursty traffic), cổng ra vẫn không bao giờ trở thành nút cổ chai gây ùn ứ ngược lại vào bên trong hệ thống.
* Bước 4: Xác định năng lực phục vụ tại cổng vào (Khu vực Mua vé)
   * Đối với Quầy vé Thường (Chiến lược Tối ưu chi phí): Áp dụng nguyên tắc dòng chảy tới hạn như các khu vực nội bộ, ta thiết lập năng lực phục vụ sát với nhu cầu thực tế để tối ưu hóa tài nguyên:$$\mu_{MuaVe\_Reg} \approx \lambda_{MuaVe\_Reg} = 70 \text{ (khách/giờ)}$$Điều này đồng nghĩa với thời gian phục vụ trung bình khoảng $0.85$ phút/khách. Tại mức này, hệ thống hoạt động ở trạng thái bão hòa, chấp nhận việc khách thường phải xếp hàng chờ đợi trong giờ cao điểm.
   * Đối với quầy vé VIP: Mặc dù lưu lượng khách VIP tại quầy rất thấp ($\lambda = 7$), ta không thiết lập $\mu = 7$ (tương đương 8.5 phút/khách). Thay vào đó, ta thiết lập:$$\mu_{MuaVe\_VIP} = 60 \text{ (khách/giờ)}$$
   Thể hiện đặc quyền VIP: Với $\lambda=7$ và $\mu=60$, hệ số sử dụng của server ($\rho$) chỉ khoảng $0.11$ ($11\%$).$$\rho_{VIP} = \frac{7}{60} \approx 11.6\% \ll 1$$Hệ số sử dụng rất thấp này đảm bảo rằng xác suất khách VIP phải chờ đợi ($P_{wait}$) là gần như bằng 0. Bất cứ khi nào khách VIP đến quầy, server hầu như luôn ở trạng thái "Rảnh" (Idle) và sẵn sàng phục vụ ngay lập tức. Đây chính là sự mô phỏng cho đặc quyền "Không cần chờ đợi" của vé VIP, trái ngược với cảnh xếp hàng tại quầy vé Thường.
   
* Bên cạnh đó, để tránh trường hợp khách đi lại vô tận trong hệ thống mà không thoát ra ngoài, thì nhóm em có sử dụng cơ chế "Mô phỏng sự mệt mỏi" (Fatigue Mechanism).
  * Cơ chế này nhằm phản ánh hành vi thực tế của khách hàng: sau khi tham gia một số lượng dịch vụ nhất định, sức khỏe và sự hứng thú sẽ giảm dần, dẫn đến xu hướng muốn rời khỏi khu vui chơi (Check-out) tăng lên, thay vì tiếp tục di chuyển ngẫu nhiên theo xác suất cố định ban đầu.
  * Cụ thể, xác suất chuyển đến trạng thái CheckOut của một khách hàng thứ $i$ tại thời điểm $t$ sẽ không còn là hằng số, mà được cập nhật động dựa trên lịch sử di chuyển của khách hàng đó theo công thức sau:$$P_{CheckOut}^{(new)} = \min\left(1.0, \quad P_{CheckOut}^{(base)} + P_{extra}\right)$$Trong đó, $P_{extra}$ (xác suất cộng thêm do mệt mỏi) được tính toán như sau:$$P_{extra} = \begin{cases}
  0 & \text{nếu } N < N_{th} \\
  \alpha \times (N - N_{th} + 1) & \text{nếu } N \ge N_{th}
  \end{cases}$$
  Các tham số trong mô hình được định nghĩa:
    * $N$ (total_visits): Tổng số lần khách hàng đã đi qua các trạm dịch vụ (Service Nodes) tính đến thời điểm hiện tại. Biến này đại diện cho độ dài hành trình tích lũy của khách.
    * $N_{th}$ (fatigue_threshold): Ngưỡng mệt mỏi. Đây là giới hạn mềm về số lượng dịch vụ mà một khách hàng trung bình có thể trải nghiệm trước khi bắt đầu cảm thấy mệt. Trong mô phỏng này, nhóm thiết lập $N_{th} = 3$.
    * $α$ (bonus_per_visit): Hệ số phạt (Fatigue Penalty Factor). Đây là trọng số xác suất được cộng dồn vào xu hướng rời đi sau mỗi lần vượt ngưỡng. Nhóm thiết lập $\alpha = 18$.
  Cơ chế hoạt động:Khi khách hàng chưa đạt đến ngưỡng mệt mỏi ($N < N_{th}$), hệ thống sử dụng ma trận xác suất chuyển tiếp gốc (Base Transition Matrix).Khi khách hàng vượt ngưỡng ($N \ge N_{th}$), xác suất CheckOut sẽ tăng tuyến tính theo cấp số cộng với công sai $\alpha$. Ví dụ: Tại lần ghé thăm thứ 4 (vượt ngưỡng 1 lần), xác suất về tăng thêm 18%; tại lần thứ 5, tăng thêm 36%.Để đảm bảo tính hợp lệ toán học (tổng xác suất tại một nút luôn bằng 1), khi $P_{CheckOut}$ tăng lên, xác suất di chuyển sang các khu vực khác (Trò chơi, Ăn uống...) sẽ được giảm xuống theo tỷ lệ tương ứng (Normalization).

````python
import simpy
import numpy as np
import csv
from enum import Enum
import pandas as pd
import sys
import io

# ========================================
# 1. ENUM & CẤU HÌNH
# ========================================

class CustomerType(Enum):
    REG = "REG"
    VIP = "VIP"

# Cấu hình khu vực: (số server, thời gian phục vụ TB (giờ), ưu tiên VIP, chỉ VIP)
AREAS_CONFIG = {
    "MuaVe_Thuong":    (2, 1/70, False, False),
    "MuaVe_VIP":       (1, 1/60, False, True),
    "CheckIn":         (2, 1/110, True,  False),
    "TroChoi_Thuong":  (2, 1/97, False, False),
    "TroChoi_VIP":     (1, 1/33, False, True),
    "TrienLam":        (1, 1/32, False, True),
    "AnUong":          (2, 1/81, False, False),
    "CheckOut":        (1, 1/200, False, False),
}

# Danh sách khu vực chính (khách có thể tự do di chuyển)
MAIN_AREAS = ["TroChoi_Thuong", "TroChoi_VIP", "TrienLam", "AnUong"]

# Xác suất chuyển tiếp (sẽ được tự động điền)
TRANSITION_PROBS = {
    "CheckIn": {
        CustomerType.REG: {
            "TroChoi_Thuong": 0.70,
            "AnUong":         0.30,
            "CheckOut":       0.00,
        },
        CustomerType.VIP: {
            "TroChoi_VIP":    0.50,
            "TrienLam":       0.30,
            "AnUong":         0.20,
            "CheckOut":       0.00,
        }
    },
    "TroChoi_Thuong": {CustomerType.REG: {}},
    "TroChoi_VIP":    {CustomerType.VIP: {}},
    "TrienLam":       {CustomerType.VIP: {}},
    "AnUong":         {CustomerType.REG: {}, CustomerType.VIP: {}},
    "MuaVe_Thuong":   {CustomerType.REG: {"CheckIn": 1.0}},
    "MuaVe_VIP":      {CustomerType.VIP: {"CheckIn": 1.0}},
}

def build_dynamic_transitions():
    for area in MAIN_AREAS:
        possible_next = ["CheckOut"]  # Luôn có thể ra về
        # REG: chỉ đi TroChoi_Thuong + AnUong
        if area in ["TroChoi_Thuong", "AnUong"]:
            possible_next.extend(["TroChoi_Thuong", "AnUong"])
        # VIP: đi TroChoi_VIP + TrienLam + AnUong
        if area in ["TroChoi_VIP", "TrienLam", "AnUong"]:
            possible_next.extend(["TroChoi_VIP", "TrienLam", "AnUong"])
        possible_next = list(set(possible_next))  # Loại trùng
        n = len(possible_next)
        probs = {a: 1.0 / n for a in possible_next}
        # Gán cho đúng loại khách
        if area == "TroChoi_Thuong":
            TRANSITION_PROBS[area][CustomerType.REG] = probs
        elif area == "TroChoi_VIP":
            TRANSITION_PROBS[area][CustomerType.VIP] = probs
        elif area == "TrienLam":
            TRANSITION_PROBS[area][CustomerType.VIP] = probs
        else:  # AnUong
            TRANSITION_PROBS[area][CustomerType.REG] = probs.copy()
            vip_probs = probs.copy()
            vip_probs["CheckOut"] += 0.1
            total = sum(vip_probs.values())
            vip_probs = {k: v/total for k, v in vip_probs.items()}
            TRANSITION_PROBS[area][CustomerType.VIP] = vip_probs

# GỌI HÀM XÂY DỰNG XÁC SUẤT
build_dynamic_transitions()

# ========================================
# 2. LOGGER (CSV + Console)
# ========================================

class Logger:
    def __init__(self, filename="simpy_simulation_log.csv"):
        self.filename = filename
        self.start_time = None
        with open(filename, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['Rel_Time', 'Abs_Time', 'Event', 'Cust_ID', 'Type', 'Area', 'Server', 'Queue', 'Details'])

    def log(self, env, event, customer, area_name=None, server=None, queue_len=None, details=""):
        if self.start_time is None:
            self.start_time = env.now
        rel_time = round(env.now - self.start_time, 5)
        abs_time = round(env.now, 5)

        with open(self.filename, 'a', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow([
                f"{rel_time:.4f}", f"{abs_time:.4f}", event,
                customer.get('id', '-'), customer.get('type').value if customer.get('type') is not None else "-",
                area_name or "-", server if server is not None else "-",
                queue_len if queue_len is not None else 0,
                details
            ])

        server_str = f"S{server}" if server is not None else "-"
        area_str = area_name or "-"
        queue_str = queue_len if queue_len is not None else "-"
        print(f"[t={rel_time:8.5f}] {event:15} | C{customer.get('id', '-'):3} ({customer.get('type').value if customer.get('type') is not None else '-'}) | {area_str:15} | {server_str} | Q={queue_str:2} | {details}")

    # --------- LOG INSIDE_STATE ---------
    def log_inside(self, env, inside, event, customer):
        rel_time = round(env.now, 5)
        abs_time = round(env.now, 5)

        with open(self.filename, 'a', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow([
                f"{rel_time:.5f}",
                f"{abs_time:.5f}",
                event,
                customer.get('id', '-'),
                customer.get('type').value if customer.get('type') else "-",
                "INSIDE_STATE",
                "-",
                inside,
                f"Inside={inside}"
            ])
        print(f"[t={rel_time:8.5f}] {event:15} | INSIDE={inside}")

# ========================================
# 3. TẠO ĐƯỜNG ĐI ĐỘNG
# ========================================

def generate_dynamic_path(customer_type, is_online, fatigue_threshold=3, bonus_per_visit=0.18):
    path = []
    visit_count = {area: 0 for area in AREAS_CONFIG.keys()}
    current = "CheckIn" if is_online else ("MuaVe_VIP" if customer_type == CustomerType.VIP else "MuaVe_Thuong")
    path.append(current)
    visit_count[current] += 1

    while current != "CheckOut":
        probs_dict = TRANSITION_PROBS.get(current, {}).get(customer_type, {}).copy()
        if not probs_dict:
            break

        total_visits = sum(visit_count.values())
        if total_visits >= fatigue_threshold:
            extra = bonus_per_visit * (total_visits - fatigue_threshold + 1)
            checkout_prob = min(1.0, probs_dict.get("CheckOut", 0.0) + extra)
            probs_dict["CheckOut"] = checkout_prob

            remaining = 1.0 - checkout_prob
            if remaining > 0:
                others = {k: v for k, v in probs_dict.items() if k != "CheckOut"}
                total_others = sum(others.values())
                if total_others > 0:
                    scale = remaining / total_others
                    for k in others:
                        probs_dict[k] *= scale

        areas = list(probs_dict.keys())
        weights = list(probs_dict.values())
        weights = np.array(weights)
        if weights.sum() == 0:
            weights = np.ones(len(weights))
        weights = weights / weights.sum()

        current = np.random.choice(areas, p=weights)
        path.append(current)
        visit_count[current] += 1

        if len(path) > 100:
            path.append("CheckOut")
            break

    if path[-1] != "CheckOut":
        path.append("CheckOut")

    return path

# ========================================
# 4. HÀNH VI KHÁCH
# ========================================

def customer_process(env, customer, path, logger, stats, resources,inside,event):
    customer['arrival_time'] = env.now
    customer['service_times'] = []
    customer['last_time'] = env.now
    logger.log(env, "ARRIVAL_SYSTEM", customer, details=f"Path: {' → '.join(path)}")
    logger.log_inside(env, inside, event, customer)
    for area_name in path:
        res = resources[area_name]

        # BỎ QUA KHU CHỈ DÀNH CHO VIP
        if res.vip_only and customer['type'] == CustomerType.REG:
            logger.log(env, "SKIPPED", customer, area_name, details="REG không được vào khu VIP")
            continue

        queue_before = len(res.queue) + len(res.users)
        priority = 0 if (customer['type'] == CustomerType.VIP and res.vip_priority) else 1

        with res.request(priority=priority) as req:
            if queue_before == 0:
                customer['last_time'] = env.now
                logger.log(env, "FIRST_IN", customer, area_name, queue_len=1,
                          details=f"Priority={priority}")
            else:
                customer['last_time'] = env.now
                logger.log(env, "QUEUE_JOIN", customer, area_name, queue_len=queue_before + 1,
                          details=f"Priority={priority}")
            yield req

            queue_after = len(res.queue) + len(res.users)
            server_id = res.users.index(req) if req in res.users else -1

            wait_time = env.now - customer['last_time']
            #customer['last_time'] = env.now

            logger.log(env, "SERVICE_START", customer, area_name, server=server_id,
                      queue_len=queue_after, details=f"Wait: {wait_time:.5f}h")

            # Vẫn dùng np.random.exponential (API cũ) -> dùng chung seed 42
            service_time = np.random.exponential(res.service_time_mean)
            customer['service_times'].append(service_time)
            yield env.timeout(service_time)

            logger.log(env, "SERVICE_END", customer, area_name, server=server_id,queue_len=len(res.queue)+ len(res.users)-1,
                      details=f"Served: {service_time:.5f}h")
            #customer['last_time'] = env.now #

    # Check-out
    total_time = env.now - customer['arrival_time']
    total_service = sum(customer['service_times'])
    wait_total = total_time - total_service

    stats['total_times'][customer['type']].append(total_time)
    stats['wait_times'][customer['type']].append(wait_total)

    logger.log(env, "CHECKOUT", customer, details=f"Total: {total_time:.5f}h | Wait: {wait_total:.5f}h")
    # GIẢI PHÓNG SLOT
    stats['inside'] = max(0, stats.get('inside', 1) - 1)
    logger.log_inside(env, stats['inside'], "INSIDE_EXIT", customer)

# ========================================
# 5. TẠO KHÁCH (SỬ DỤNG RNG ĐỘC LẬP CHO CÁC SỰ KIỆN LIÊN QUAN ĐẾN THỜI GIAN ĐẾN)
# ========================================

PARK_CAPACITY = 200

# THAY ĐỔI: Thêm rng_arrival làm tham số
def customer_generator(env, lambda_reg, lambda_vip, logger, stats, resources, rng_arrival):
    cust_id = 0
    total_lambda = lambda_reg + lambda_vip
    vip_prob = lambda_vip / total_lambda if total_lambda > 0 else 0

    # 30% REG online, 30% VIP online
    online_prob_reg = 0.3
    online_prob_vip = 0.3

    while True:
        # Xác định VIP / REG (DÙNG RNG ĐỘC LẬP)
        is_vip = rng_arrival.random() < vip_prob
        ctype = CustomerType.VIP if is_vip else CustomerType.REG

        # Xác định online theo loại khách (DÙNG RNG ĐỘC LẬP)
        if ctype == CustomerType.VIP:
            is_online = rng_arrival.random() < online_prob_vip
        else:  # REG
            is_online = rng_arrival.random() < online_prob_reg

        # Sinh path dựa trên loại khách + online (Vẫn dùng np.random API cũ)
        path = generate_dynamic_path(ctype, is_online,
                                     fatigue_threshold=3,
                                     bonus_per_visit=0.18)

        # Thông tin khách
        customer = {'id': cust_id, 'type': ctype, 'is_online': is_online}
        cust_id += 1
        stats['arrived'][ctype] += 1

        # Xử lý vào công viên (ENTRY)
        first_area = path[0] if path else None
        entry_points = ["MuaVe_Thuong", "MuaVe_VIP", "CheckIn"]

        if first_area in entry_points:
            if stats.get('inside', 0) >= PARK_CAPACITY:
                logger.log(env, "REJECTED", customer,
                           details=f"FULL SLOT ({stats.get('inside',0)}/{PARK_CAPACITY}) - Không được mua vé")
                logger.log_inside(env, stats['inside'], "INSIDE_REJECT", customer)

                # REJECTED: Vẫn tiêu thụ số ngẫu nhiên TỪ RNG ĐỘC LẬP
                yield env.timeout(rng_arrival.exponential(1 / total_lambda))
                continue # Số ngẫu nhiên đã được tiêu thụ, nhưng khách bị bỏ qua
            else:
                stats['inside'] = stats.get('inside', 0) + 1
                if ctype == CustomerType.REG:
                    if is_online:
                        stats['reg_online'] += 1
                    else:
                        stats['reg_offline'] += 1
                else:
                    if is_online:
                        stats['vip_online'] += 1
                    else:
                        stats['vip_offline'] += 1
                # logger.log_inside(env, stats['inside'], "INSIDE_ENTER", customer)

        # Log đến nơi đầu tiên
        logger.log(env, "ARRIVAL", customer, details=f"→ {path[0]} (online={is_online})")

        yield env.timeout(0)

        # Chạy process mô phỏng khách di chuyển
        env.process(customer_process(env, customer, path, logger, stats, resources,stats['inside'],"INSIDE_ENTER"))

        # Log đã tạo xong customer
        logger.log(env, "GENERATED", customer, details=f"→ {path[0]} (online={is_online})")

        # Thời gian giữa các khách đến: DÙNG RNG ĐỘC LẬP
        yield env.timeout(rng_arrival.exponential(1 / total_lambda))

# ========================================
# 6. PHÂN TÍCH HÀNG ĐỢI THỰC NGHIỆM (GIỮ NGUYÊN)
# ========================================
import re

def empirical_queue_analysis(log_file, duration):
    df = pd.read_csv(log_file)

    # Chuyển đổi dữ liệu số
    df['Rel_Time'] = pd.to_numeric(df['Rel_Time'], errors='coerce')
    if 'Queue' in df.columns:
        df['Queue'] = pd.to_numeric(df['Queue'], errors='coerce').fillna(0)
    else:
        df['Queue'] = 0

    # --- 1. TỔNG QUAN (GIỮ NGUYÊN NHƯ CŨ) ---
    service_starts = df[df['Event'] == 'SERVICE_START']
    arrival_rate = service_starts['Area'].value_counts() / duration

    wait_pattern = re.compile(r'Wait:\s*([\d.]+)h')
    length_pattern = re.compile(r'Q=\s*(\d+)')
    waits_by_area = {}
    for _, row in service_starts.iterrows():
        area = row['Area']
        details = str(row.get('Details', ''))
        m = wait_pattern.search(details)
        if m:
            waits_by_area.setdefault(area, []).append(float(m.group(1)))

    timeavg_Lq = {}
    sample_mean_queue = {}
    areas = [k for k in AREAS_CONFIG.keys()]

    for area in areas:
        df_area = df[df['Area'] == area].sort_values('Rel_Time')
        if df_area.empty:
            timeavg_Lq[area] = 0.0; sample_mean_queue[area] = 0.0
            continue

        t_start = float(df_area['Rel_Time'].iloc[0])
        t_end = max(float(df_area['Rel_Time'].iloc[-1]), float(duration))
        times = df_area['Rel_Time'].astype(float).values
        queues = df_area['Queue'].astype(float).values

        last_t, last_q, q_time_area = times[0], queues[0], 0.0
        for t, q in zip(times[1:], queues[1:]):
            q_time_area += last_q * max(0, t - last_t)
            last_t, last_q = t, q
        q_time_area += last_q * max(0, t_end - last_t) # Đoạn cuối

        timeavg_Lq[area] = q_time_area / (t_end - t_start) if t_end > t_start else 0
        sample_mean_queue[area] = queues.mean() if len(queues) > 0 else 0

    # --- IN BẢNG TỔNG QUAN ---
    print(f"\n{'='*100}")
    print(f"PHÂN TÍCH HÀNG ĐỢI THỰC NGHIỆM (TỔNG HỢP) - {duration}h")
    print(f"{'='*100}")
    print(f"{'Khu vực':15} {'λ(/h)':>7} {'μ(/h)':>8} {'c':>3} {'ρ':>6} {'Lq_timeavg':>12} {'Lq_sample':>10} {'Wq (phút)':>12}")
    print(f"{'-'*100}")

    for area in AREAS_CONFIG:
        c, mean_time, _, _ = AREAS_CONFIG[area]
        mu = 1 / mean_time if mean_time > 0 else 0.0
        lambda_ = arrival_rate.get(area, 0.0)
        rho = (lambda_ / (c * mu)) if (c * mu) > 0 else 0.0
        lq_ta = timeavg_Lq.get(area, 0.0)
        lq_sm = sample_mean_queue.get(area, 0.0)

        waits = waits_by_area.get(area, [])
        wq_min = (np.mean(waits) * 60) if waits else 0.0

        print(f"{area:15} {lambda_:7.2f} {mu:8.2f} {c:3d} {rho:6.3f} {lq_ta:12.3f} {lq_sm:10.3f} {wq_min:12.2f}")
    print(f"{'='*100}")
    print(f"\n{'='*100}")
    print("CHI TIẾT KHU VỰC CHECKIN (REG vs VIP)")
    print(f"{'='*100}")

    # Lấy dữ liệu CheckIn và sắp xếp chuẩn theo thời gian thực
    checkin_df = df[df['Area'] == 'CheckIn'].sort_values('Rel_Time')

    # --- Tính Lambda & Wq riêng cho từng loại (Giữ nguyên phần này vì đúng) ---
    starts = checkin_df[checkin_df['Event'] == 'SERVICE_START']

    checkin_stats = {}
    for ctype in ['REG', 'VIP']:
        type_data = starts[starts['Type'] == ctype]
        count = len(type_data)
        lam = count / duration

        w_list = []
        for det in type_data['Details']:
            m = wait_pattern.search(str(det))
            if m: w_list.append(float(m.group(1)))

        wq_avg_h = np.mean(w_list) if w_list else 0.0
        checkin_stats[ctype] = {'lambda': lam, 'Wq_min': wq_avg_h * 60}


    # Lấy Lq tổng của CheckIn (đã tính chính xác ở phần Tổng quan)
    lq_total_checkin = timeavg_Lq.get('CheckIn', 0.0)

    print(f"{'Loại Khách':15} {'λ (khách/h)':>15} {'Wq (phút)':>15}")
    print("-" * 65)
    print(f"{'REG':15} {checkin_stats['REG']['lambda']:15.2f} {checkin_stats['REG']['Wq_min']:15.2f} ")
    print(f"{'VIP':15} {checkin_stats['VIP']['lambda']:15.2f} {checkin_stats['VIP']['Wq_min']:15.2f} ")
    print("-" * 65)


    # THỜI GIAN ĐÁP ỨNG E[R] = Wq + E[S] CHO CÁC LOẠI KHÁCH
    print(f"\n{'='*100}")
    print("THỜI GIAN ĐÁP ỨNG TRUNG BÌNH E[R] = Wq + E[S] (Thực tế, đơn vị phút)")
    print(f"{'='*100}")

    # Lấy dữ liệu Wq và E[S] cho tất cả khu vực/loại
    # 1. Wq
    w_df = df[df['Event'] == 'SERVICE_START'].copy()
    w_df['W_val'] = w_df['Details'].str.extract(r'Wait: ([\d.]+)').astype(float)
    avg_wq_group = w_df.groupby(['Area', 'Type'])['W_val'].mean()

    # 2. E[S] (Service Time thực tế)
    s_df = df[df['Event'] == 'SERVICE_END'].copy()
    s_df['S_val'] = s_df['Details'].str.extract(r'Served: ([\d.]+)').astype(float)
    avg_es_group = s_df.groupby(['Area', 'Type'])['S_val'].mean()

    # 3. Định nghĩa lộ trình
    reg_route = ["MuaVe_Thuong", "CheckIn", "TroChoi_Thuong", "AnUong", "CheckOut"]
    vip_route = ["MuaVe_VIP", "CheckIn", "TroChoi_VIP", "TrienLam", "AnUong", "CheckOut"]

    def print_route_stats(route_name, route_areas, ctype):
        print(f"--- KHÁCH {ctype} ({route_name}) ---")
        print(f"{'Khu vực':20} {'Wq':>10} {'+':>3} {'E[S]':>10} {'=':>3} {'E[R] (phút)':>12}")
        print("-" * 65)
        total_R = 0

        for area in route_areas:
            # Lấy Wq (giờ)
            if (area, ctype) in avg_wq_group:
                wq_h = avg_wq_group[area, ctype]
            else:
                wq_h = 0.0

            # Lấy E[S] (giờ) - nếu không có log service end (do ít khách), lấy lý thuyết
            if (area, ctype) in avg_es_group:
                es_h = avg_es_group[area, ctype]
            else:
                # Fallback lý thuyết nếu chưa có mẫu nào
                if area in AREAS_CONFIG:
                    es_h = AREAS_CONFIG[area][1]
                else:
                    es_h = 0.0

            er_h = wq_h + es_h
            total_R += er_h

            print(f"{area:20} {wq_h*60:10.2f} {'+':>3} {es_h*60:10.2f} {'=':>3} {er_h*60:12.2f}")

        print("-" * 65)
        print(f"{'TỔNG CỘNG':20} {' ':10} {' ':3} {' ':10} {' ':3} {total_R*60:12.2f}")
        print()

    print_route_stats("Lộ trình Thường", reg_route, "REG")
    print_route_stats("Lộ trình VIP", vip_route, "VIP")
    print(f"{'='*100}\n")
# ========================================
# 7.run_simulation
# ========================================

# THAY ĐỔI: Thêm rng_arrival làm tham số với giá trị mặc định (nếu không được truyền)
def run_simulation(duration=50.0, lambda_reg=100, lambda_vip=10, rng_arrival=np.random.default_rng(42)):
    env = simpy.Environment()
    logger = Logger()
    stats = {
        'wait_times': {CustomerType.VIP: [], CustomerType.REG: []},
        'total_times': {CustomerType.VIP: [], CustomerType.REG: []},
        'arrived': {CustomerType.VIP: 0, CustomerType.REG: 0},
        'inside': 0,
        'reg_online': 0, 'reg_offline': 0,
        'vip_online': 0, 'vip_offline': 0
    }
    resources = {}
    for name, (capacity, mean_time, vip_priority, vip_only) in AREAS_CONFIG.items():
        res = simpy.PriorityResource(env, capacity=capacity)
        res.service_time_mean = mean_time
        res.vip_priority = vip_priority
        res.vip_only = vip_only
        resources[name] = res

    # THAY ĐỔI: Truyền rng_arrival vào customer_generator
    env.process(customer_generator(env, lambda_reg, lambda_vip, logger, stats, resources, rng_arrival))

    print(f"\nBẮT ĐẦU MÔ PHỎNG SIMPY ({duration} GIỜ)")
    print("Log: thời gian, sự kiện, khách, khu vực, server, hàng đợi, chi tiết\n")
    env.run(until=duration)

    print("\n" + "="*70)
    print("MÔ PHỎNG HOÀN TẤT")
    print(f"Log file: {logger.filename}")
    print("="*70)

    # --- THỐNG KÊ KHÁCH ---
    n_reg_complete = len(stats['total_times'][CustomerType.REG])
    n_vip_complete = len(stats['total_times'][CustomerType.VIP])
    n_reg_arrived = stats['arrived'][CustomerType.REG]
    n_vip_arrived = stats['arrived'][CustomerType.VIP]

    print(f"\nTỔNG KHÁCH ĐẾN:")
    print(f"  REG: {n_reg_arrived} khách")
    print(f"  VIP: {n_vip_arrived} khách")

    print(f"\nTỔNG KHÁCH HOÀN THÀNH:")
    print(f"  REG: {n_reg_complete} khách")
    print(f"  VIP: {n_vip_complete} khách")

    print(f"\nPHÂN LOẠI MUA VÉ (VÀO ĐƯỢC):")
    print(f"  REG Online: {stats['reg_online']}")
    print(f"  REG Offline: {stats['reg_offline']}")
    print(f"  VIP Online: {stats['vip_online']}")
    print(f"  VIP Offline: {stats['vip_offline']}")

    # --- GỌI HÀM PHÂN TÍCH THỰC NGHIỆM ---
    empirical_queue_analysis(logger.filename, duration)

# --- HÀM BẮT ĐẦU RA CONSOLE ---

# THAY ĐỔI: Thêm rng_arrival làm tham số
def capture_and_run_simulation(duration, lambda_reg, lambda_vip, rng_arrival):
    old_stdout = sys.stdout
    redirected_output = io.StringIO()
    sys.stdout = redirected_output

    try:
        # Gọi hàm chạy mô phỏng gốc, truyền rng_arrival
        run_simulation(duration=duration, lambda_reg=lambda_reg, lambda_vip=lambda_vip, rng_arrival=rng_arrival)

    finally:
        # KHÔI PHỤC đầu ra console gốc
        sys.stdout = old_stdout

    # Trả về toàn bộ nội dung đã được print
    return redirected_output.getvalue()
# # ========================================
# # 8. CHẠY 1 TRƯỜNG, TỶ LỆ VIP:REG = 1:10
# # ========================================
# if __name__ == "__main__":

#     # BƯỚC 1: TẠO BỘ TẠO NGẪU NHIÊN ĐỘC LẬP CHO CÁC SỰ KIỆN ARRIVAL
#     # Điều này đảm bảo số lượng khách TỔNG ĐẾN cố định (fixed input stream)
#     rng_arrival = np.random.default_rng(42)

#     # BƯỚC 2: ĐẶT SEED CHUNG CHO CÁC SỰ KIỆN PHỤC VỤ (SERVICE TIME, PATH)
#     # Điều này đảm bảo thời gian phục vụ và đường đi của khách là cố định (fixed internal process)
#     np.random.seed(42)

#     lambda_reg_values = [150]
#     vip_reg_ratio = 1 / 10

#     # Biến lưu trữ TẤT CẢ đầu ra console từ nhiều lần chạy
#     full_console_output = ""

#     print("BẮT ĐẦU NGHIÊN CỨU: λ_reg → TỶ LỆ VIP:REG = 1:10")
#     print("="*90)

#     for λ_reg in lambda_reg_values:
#         # RESET rng_arrival và np.random.seed(42) VỚI CÙNG SEED
#         # Note: np.random.seed(42) đã được gọi 1 lần ở trên, không cần gọi lại trong vòng lặp này vì chỉ có 1 λ_reg.
#         # Nếu có nhiều λ_reg, bạn nên gọi lại np.random.seed(42) VÀ tạo lại rng_arrival ở đây.

#         λ_vip = λ_reg * vip_reg_ratio

#         # Bắt đầu log cho lần chạy hiện tại
#         header = f"\n{'='*35} CHẠY λ_reg = {λ_reg}, λ_vip = {λ_vip:.1f} {'='*35}\n"
#         print(header)
#         full_console_output += header

#         # CHẠY VÀ BẮT ĐẦU RA
#         # THAY ĐỔI: Truyền rng_arrival vào hàm
#         current_run_output = capture_and_run_simulation(
#             duration=100.0,
#             lambda_reg=λ_reg,
#             lambda_vip=λ_vip,
#             rng_arrival=rng_arrival
#         )

#         # In ra màn hình Colab (để xem log trực tiếp)
#         print(current_run_output)

#         # Lưu vào biến tổng để tải về
#         full_console_output += current_run_output
#         full_console_output += "\n" + "="*100 + "\n"

#     # ==========================================================
#     # LƯU VÀ TẢI VỀ TẤT CẢ OUTPUT CONSOLE (LOG + PHÂN TÍCH)
#     # ==========================================================
#     from google.colab import files

#     output_filename = "trung60.txt"
#     with open(output_filename, "w", encoding='utf-8') as f:
#         f.write(full_console_output)

#     print("\n\n=== MÔ PHỎNG VÀ PHÂN TÍCH HOÀN TẤT ===")
#     print(f"File log console chứa toàn bộ bảng phân tích đã được tạo: {output_filename}")
#     #files.download(output_filename)
````

# 5.Kết luận
## 5.1. Tổng kết công việc
Nhóm Desert Serpent đã hoàn thành việc thiết kế và mô phỏng hệ thống hàng đợi cho Khu vui chơi giải trí, đi từ mô hình lý thuyết (Phase 1) đến hiện thực hóa bằng mã nguồn Python/SimPy (Phase 2).Với vai trò là Project Coordinator, tôi xác nhận nhóm đã thực hiện thành công các hạng mục sau:Xây dựng mô hình mạng lưới hàng đợi (Jackson Network) với phân luồng rõ ràng giữa khách VIP và khách Thường.Hiện thực hóa hệ thống mô phỏng sự kiện rời rạc (Discrete Event Simulation) để kiểm thử khả năng chịu tải của hệ thống.Thực hiện quy trình "Săn tìm điểm nghẽn" (Bottleneck Hunting) qua các kịch bản tải trọng khác nhau (60, 100, 150 khách/giờ). Kết quả là đã xác định và giải quyết triệt để các nút thắt cổ chai tại khu vực Trò chơi, Mua vé và Ăn uống, đưa thời gian chờ trung bình ($W_q$) từ mức quá tải (>60 phút) xuống mức lý tưởng (< 1 phút).
## 5.2. Điểm mạnh của giải pháp
Dựa trên kết quả thực nghiệm, hệ thống mô phỏng của nhóm đạt được những ưu điểm vượt trội sau:Quy trình tối ưu hóa dựa trên dữ liệu: Thay vì chỉ điều chỉnh cảm tính, nhóm đã áp dụng phương pháp phân tích dòng chảy tới hạn. Việc tăng số lượng máy chủ ($c$) tại các trạm Trò chơi và Ăn uống được thực hiện dựa trên bằng chứng thống kê rõ ràng từ log mô phỏng, chứng minh được hiệu quả giảm tải ngay lập tức.Mô hình hành vi khách hàng thực tế: Hệ thống không chỉ định tuyến ngẫu nhiên mà đã tích hợp cơ chế "Mô phỏng sự mệt mỏi" (Fatigue Mechanism). Việc tăng xác suất rời đi (Check-out) dựa trên số lần sử dụng dịch vụ giúp mô phỏng sát với tâm lý con người thực tế, ngăn chặn tình trạng khách hàng lặp vòng vô hạn trong hệ thống.Kiến trúc phần mềm linh hoạt: Mã nguồn được thiết kế theo hướng đối tượng (OOP) và tách biệt rõ ràng giữa Cấu hình (Config) - Xử lý (Core Simulation) - Ghi nhận (Logger). Điều này cho phép dễ dàng mở rộng thêm các khu vực mới hoặc thay đổi chính sách phục vụ (ví dụ: thay đổi quyền ưu tiên VIP) mà không ảnh hưởng đến cấu trúc cốt lõi.
##5.3. Các hạn chế tồn tại (Limitations)
Theo đánh giá từ thành viên kiểm thử (Testing member), mô hình hiện tại vẫn tồn tại 3 hạn chế cần cải thiện trong tương lai:Thiếu hành vi "Balking" và "Reneging" (Bỏ cuộc):Hiện tại, khách hàng trong mô phỏng vẫn kiên nhẫn xếp hàng dù hàng đợi rất dài (ví dụ: đợi 69 phút tại khu Trò chơi trong kịch bản quá tải).Thực tế: Khách hàng thường sẽ bỏ sang khu vực khác nếu thấy hàng quá dài (Balking) hoặc rời hàng khi đã đợi quá lâu (Reneging). Mô hình chưa phản ánh được hành vi tâm lý này.Giả định phân phối thời gian phục vụ đơn giản:Mô hình đang sử dụng phân phối Mũ (Exponential Distribution) cho toàn bộ các trạm phục vụ ($M/M/c$).Thực tế: Các trò chơi cảm giác mạnh thường có thời gian cố định (Deterministic) hoặc phân phối Chuẩn (Normal). Việc dùng phân phối Mũ có thể tạo ra độ biến thiên lớn hơn thực tế đôi chút.Chưa xử lý giai đoạn khởi động (Warm-up Period):Kết quả thống kê hiện tại tính trung bình trên toàn bộ thời gian chạy (0 - 100h), bao gồm cả giai đoạn đầu khi công viên vừa mở cửa và chưa có khách.Hệ quả: Điều này có thể làm các chỉ số trung bình ($L_q, W_q$) thấp hơn một chút so với thực tế khi hệ thống đã đi vào trạng thái ổn định (Steady state). Cần loại bỏ dữ liệu của khoảng 10-15% thời gian đầu để có kết quả chính xác hơn.
